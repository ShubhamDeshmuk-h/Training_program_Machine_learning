```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('wisc_bc_data.csv')
```

```python
df.head()
```

## seperate the data

```python
x=df.drop(['id','diagnosis'],axis =1)
y=df['diagnosis']
```

```python
## standardisation is preferred for ANN than normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

## cross validation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=0,stratify=y)
```

## build our model

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf=MLPClassifier(hidden_layer_sizes=(64,128,128),activation='relu',max_iter=50, verbose= True)## 3 hidden layers with 15 neurons eahc
```

```python
clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
