```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wisc_bc_data.csv')
```

```python
df
```

```python
x = df.drop(['id','diagnosis'],axis = 1)
```

```python
y = df['diagnosis']
```

# normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sca = MinMaxScaler()
x_scaled = sca.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca;
```

```python
var  = pca.explained_variance_ratio_
```

```python
x = x.columns
```

```python
plt.figure(figsize = (36,25))
plt.title('ALL VS diagnosis')
plt.xlabel('All')
plt.ylabel('Variance')
plt.bar(x,var,edgecolor = 'black', width = 0.6,linewidth = 3);
plt.grid();
```

```python
plt.figure(figsize = (16,29))
plt.title('vVariance caused by features')
plt.xlabel('features')
plt.ylabel('Variance')
plt.xticks(range(1,31,1))
plt.bar(range(1,31),var,edgecolor = 'black', width = 0.6,color = 'darkblue');
plt.grid();
```

```python
pca = PCA(n_components = 5)
```

```python
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x_pca,y,random_state = 0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'rbf')
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_pred);
```
