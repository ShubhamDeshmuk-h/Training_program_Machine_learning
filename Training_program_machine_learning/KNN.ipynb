```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Training/Social_Network_Ads.csv')
df.head()
```

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()
x_scaled = sca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,
                                                 random_state = 0
    
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#how to decide how many of the neighbour we should use for the knn
# here we go for the use of the we can perform 
```

```python
np.mean(y_test != y_pred)
```

### choosine k-value

```python
error = []

for k in range(1,51):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    error.append(np.mean(y_test != y_pred))
```

```python
error
```

```python
plt.figure(figsize= (25,16))
plt.title('Error vs K')
plt.xlabel('K')
plt.ylabel('Error')
plt.xticks(range(1,51,1))
plt.plot(range(1,51),error[:51] ,color = 'black',marker = 'o')
plt.grid()
```
